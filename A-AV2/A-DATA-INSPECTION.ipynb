```python
from utils import DataFrame, plot_three_images
import numpy as np
import cv2
```

```python
def process_img(img, ystart=110, ystop=230):
    img = img[ystart:ystop, :]
    cv2.resize(img, (200, 75), cv2.INTER_AREA)
    return img
```

```python
print("=========")
red_frame = DataFrame("./road-images/red.csv")
print("=========")
green_frame = DataFrame("./road-images/green.csv")
```

```python
red_frame.distribution(my_bins=200)
```

```python
green_frame.distribution(my_bins=180)
```

```python
i, j, k = np.random.randint(red_frame.num), np.random.randint(red_frame.num), np.random.randint(red_frame.num)
```

```python
red1, red2, red3 = red_frame.get_image(i), red_frame.get_image(j), red_frame.get_image(k)
green1, green2, green3 = green_frame.get_image(i), green_frame.get_image(j), green_frame.get_image(k)
```

```python
plot_three_images(red1, red2, red3, w =17, h=5)
plot_three_images(green1, green2, green3, w =17, h=5)
```

```python
x, y, z = process_img(red1), process_img(red2), process_img(red3)
plot_three_images(x, y, z, w =17, h=5)
```

```python
x, y, z = process_img(green1), process_img(green2), process_img(green3)
plot_three_images(x, y, z, w =17, h=5)
```
