```python
from helpers import BirdsEye, plot_three_images, make_img_list, make_cropped_list, make_debug_imgs
```

```python
# (x, y) lowerleft, upperleft, upperright, lowerright
SOURCEPOINTS = [(0, 43), (130, 15), (190, 15), (320, 43)]
DESTINATIONPOINTS = [(130, 50), (130, 10), (190, 10), (190, 50)]
BIRDSEYE = BirdsEye(SOURCEPOINTS, DESTINATIONPOINTS)

PATH = "./samples/tests-perspectives-images/"
YSTART, YSTOP = 100, 155
```

```python
def helper_process_image_sets(my_prefix, my_number_of_imgs):
    imgs = make_img_list(PATH, prefix=my_prefix, suffix=".jpg", number_of_images=my_number_of_imgs)
    cropped_imgs = make_cropped_list(imgs, ystart=YSTART, ystop=YSTOP)
    raw_imgs, sky_imgs = make_debug_imgs(cropped_imgs, BIRDSEYE)
    return imgs, raw_imgs, sky_imgs
```

```python
def helper_debug_plots(imgs, raw_imgs, sky_imgs, x, y, z):
    plot_three_images(sky_imgs[x], sky_imgs[y], sky_imgs[z], w=17, h=5)
    plot_three_images(raw_imgs[x], raw_imgs[y], raw_imgs[z], w=17, h=5)
    plot_three_images(imgs[x], imgs[y], imgs[z], w=17, h=5)
```

```python
imgs1, raw_imgs1, sky_imgs1 = helper_process_image_sets(my_prefix="img", my_number_of_imgs=6)
imgs2, raw_imgs2, sky_imgs2 = helper_process_image_sets(my_prefix="i", my_number_of_imgs=15)
```

```python
helper_debug_plots(imgs1, raw_imgs1, sky_imgs1, 0, 1, 2)
```
